***<span style="font-size:20px;">Outlier Detection</span>***

```python
selected_col=['mean radius', 'mean perimeter', 'mean area', 'mean concavity',
       'mean concave points', 'worst radius', 'worst perimeter', 'worst area',
       'worst concavity', 'worst concave points']
```

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(selected_col):
    plt.subplot((len(selected_col) + 2) // 3, 3, i + 1)
    sns.boxplot(x[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
for i in selected_col:
    q1=x[i].quantile(0.25)
    q3=x[i].quantile(0.75)
    iqr=q3-q1
    lower_bound=q1-(1.5*iqr)
    upper_bound=q3+(1.5*iqr)
    df=x[(x[i]>lower_bound) & (x[i]<upper_bound)]
```

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(selected_col):
    plt.subplot((len(selected_col) + 2) // 3, 3, i + 1)
    sns.boxplot(df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

**Even after iqr method there are outlier**

```python
skew_values = x[selected_col].skew()
print("Skewness for Selected Columns:")
print(skew_values)
```

*We can see that most of the columns are highly skewed that is why outlier elimination using IQR didn't work properly*

**Lets use Logarithmic Transformation to handle skewness**

```python
df1=x.copy()
for col in selected_col:
    df1[col] = np.log1p(df1[col])  # Use log1p to handle zero values
```

**Skwe valus after Logarithmic Transformation**

```python
skew_values = df1[selected_col].skew()
print("Skewness for Selected Columns:")
print(skew_values)
```

Now we can see that skew values reduced significatly

**Box plot after adjusting skewness**

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(selected_col):
    plt.subplot((len(selected_col) + 2) // 3, 3, i + 1)
    sns.boxplot(df1[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

Now the outliers reduced significantly

**Check histogram after Logarithmic Transformation**

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(selected_col):  # Replace 'selected_col' with your list of columns
    plt.subplot((len(selected_col) + 2) // 3, 3, i + 1)
    sns.histplot(df1[col], bins=30, kde=True, color='blue')  # `df1` is your DataFrame
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Since the data is normally distributed go for Z score to eliminate outliers

**Z Score**

```python
from scipy.stats import zscore
z_scores = df1[selected_col].apply(zscore)
outliers = (z_scores.abs() > 3).any(axis=1)
print(f"Number of outliers detected: {outliers.sum()}")
```

Replace outliers with the closest non-outlier values

```python
for col in selected_col:
    upper_limit = df1[col].mean() + 3 * df1[col].std()
    lower_limit = df1[col].mean() - 3 * df1[col].std()
    df1[col] = df1[col].clip(lower=lower_limit, upper=upper_limit)
```

***FEATURE SCALIG***

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df1,y,test_size=0.2,random_state=42)
```

**Standard Scaling**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

***<span style="font-size:24px;">2. Classification Algorithm Implementation</span>***

****1. Logistic Regression****

**Logistic Regression is a statistical model used for binary classification. It calculates the probability of a data point belonging to a specific 
class using a sigmoid function. Itâ€™s suitable for this dataset because breast cancer classification is typically a binary problem, and logistic 
regression performs well with linearly separable data.**

```python
from sklearn.linear_model import LogisticRegression 
```

```python
model=LogisticRegression()
model.fit(X_train_scaled,y_train)
```

```python
y_pred=model.predict(X_test_scaled)
```

```python
y_pred
```

```python
y_test
```

```python
#Accuracy
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
```

High accuracy means model is performing well

```python
print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(3,3))
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
```
